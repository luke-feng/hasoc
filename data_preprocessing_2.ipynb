```python
import pandas as pd
```

```python
raw_data = pd.read_csv('t2_without_ge.tsv',sep='\t')
```

```python
non = raw_data.loc[raw_data['task_2'] == 'NONE']
hate = raw_data.loc[raw_data['task_2'] == 'HATE']
offn = raw_data.loc[raw_data['task_2'] == 'OFFN']
prfn = raw_data.loc[raw_data['task_2'] == 'PRFN']
```

```python
balanced_dataset=[non.sample(3351), hate.sample(3351), offn.sample(3351), prfn.sample(3351)]
balanced_dataset = pd.concat(balanced_dataset)
balanced_dataset.to_csv('balanced_dataset.tsv',sep='\t')
```
